```python
import requests
import matplotlib.pyplot as plt
import json
import time as tm
```

```python
LatList = []
LongList = []
TimeList = []
counter = 0
while counter <= 99:
    
    response = requests.get('http://api.open-notify.org/iss-now.json')
    data = response.json()

    longitude = data['iss_position']['longitude']
    latitude = data['iss_position']['longitude']
    time = data['timestamp']
    LongList.append(longitude)
    LatList.append(latitude)
    TimeList.append(time)
    tm.sleep(10)
    
    counter += 1

print(LongList)
print(LatList)
print(TimeList)
```

```python
print("The Latitudes: ", LatList)
```

```python
print("The Longitudes: ", LongList)
```

```python
print("The Timestamps: ", TimeList)
```

```python
fig, axes = plt.subplots(figsize = (100,100),dpi=100)
plt.plot(LatList, LongList,'r', lw = 2.0, linestyle = '-')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('The Path of the ISS')
plt.show()
```

```python
fig, axes = plt.subplots(figsize = (100,100),dpi=100)
plt.plot(TimeList, LatList,'g', lw = 2.0, linestyle = '-')
plt.xlabel('Time')
plt.ylabel('Latitude')
plt.title('The Latitude of the ISS')
plt.show()
```

```python
fig, axes = plt.subplots(figsize = (100,100),dpi=100)
plt.plot(TimeList, LongList,'b', lw = 2.0, linestyle = '-')
plt.xlabel('Time')
plt.ylabel('Longitude')
plt.title('The Longitude of the ISS')
plt.show()
```
